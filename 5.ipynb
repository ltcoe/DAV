```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
np.random.seed(42)
data = np.random.randn(1000)
data= np.cumsum(data)
```

```python
plt.figure(figsize=(10,6))
plt.plot(data)
plt.title('Example Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

```python
plot_acf(data,lags=20)
plt.show()
plot_pacf(data,lags=20)
plt.show()
```

```python
model = ARIMA(data, order=(2,1,1))
fit_model = model.fit()

forecast = fit_model.forecast(steps=10)
```

```python
plt.figure(figsize=(10,6))
plt.plot(data, label='Observed')
plt.plot(range(len(data),len(data)+len(forecast)),forecast,label='Forecaste', color='red')
plt.title('ARIMA Forecast')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

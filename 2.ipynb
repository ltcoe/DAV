```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
data = {'Years': [1, 2, 3, 4, 5],  # Independent variable (e.g., years of experience)
        'Salary': [30000, 35000, 40000, 45000, 50000]}  # Dependent variable (e.g., salary)

df = pd.DataFrame(data)

# Split data into features (X) and target variable (y)
X = df[['Years']]  # In simple linear regression, we only have one independent variable
y = df['Salary']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_predicted = model.predict(X_test)

# Evaluate the model (using mean squared error here)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_predicted)
print("Mean Squared Error:", mse)
```

```python

plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, model.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
